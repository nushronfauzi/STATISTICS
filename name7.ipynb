```python
import matplotlib.pyplot as plt
```

```python
P = [2,4,4,4,5,5,7,9]
Q =[10,13,15,14,13,16,18,21]
X = [1,2,3,4,5,6,7,8]
```

```python
plt.scatter(X,P, color ='r', marker='D')
plt.scatter(X,Q, color ='b', marker='*')
plt.title ('correlation betwen P and Q')
plt.xlabel('range')
plt.ylabel('value')
plt.legend ((P,Q),loc ='lower right')
plt.show()
```
